```python
# Importing NumPy Library
import numpy as np
import sys

# Reading number of unknowns
n = 3

# Making numpy array of n x n size and initializing 
# to zero for storing augmented matrix
a = np.zeros(n)

# Making numpy array of n size and initializing 
# to zero for storing solution vector
x = np.zeros(n)

# Reading augmented matrix coefficients
a = np.array([[1,4,7],[2,5,8],[3,6,11]])

print(a)

# Applying Gauss Elimination
for i in range(n):
    if a[i][i] == 0.0:
        sys.exit('Divide by zero detected!')
        
    for j in range(i+1, n):
        ratio = a[j][i]/a[i][i]
        
        for k in range(n):
            a[j][k] = a[j][k] - ratio * a[i][k]

print(a)

# Back Substitution

for i in range(n-1,-1,-1):
    print("i=",i)
    x[i] = a[i][n-1]
    
    for j in range(i+1,n):
        print("j=",j)
        x[i] = x[i] - a[i][j]*x[j]
    
    x[i] = x[i]/a[i][i]

# Displaying solution
print('\nRequired solution is: ')
for i in range(n):
    print('X%d = %0.2f' %(i,x[i]), end = '\t')
```

```python
# Applying Gauss Elimination for n*n

import numpy as np

def gauss_elimin(a):
    a = a.copy()
    n = a[0].size

    for i in range(n):
        if a[i][i] == 0.0:
            sys.exit('Divide by zero detected!')
            
        for j in range(i+1, n):
            ratio = a[j][i]/a[i][i]

            for k in range(n):
                a[j][k] = a[j][k] - ratio * a[i][k]

    return a
```

```python
# Applying Backward Substitution for n*n

from scipy.linalg import solve

def back_subs(u, b):
    u = u.copy()
    b = b.copy()
    n = b.size

    x = np.zeros((n,1))

    for k in range(n-1,-1,-1):
        x[k][0] = (b[k] - np.dot(u[k,k+1:n],x[k+1:n]))/u[k,k]

    return x
```

```python
a = np.array([[1,4,7],[2,5,8],[3,6,11]])
b = np.array([[1],[1],[1]])

# 
# a_u = np.triu(a)
a_u = gauss_elimin(a)

print(a_u)

x = back_subs(a_u, b)

# Solves the linear equation set a * x = b for the unknown x for square a matrix.
# x = solve(a_u, b)

print(x)
```
